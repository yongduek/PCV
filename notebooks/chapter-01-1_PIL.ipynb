## Image save to file

```python
filelist = ['data/aero1.jpg', 'data/aero3.jpg', 'data/apple.jpg']
```

```python
import os
```

```python
for infile in filelist:
    splittextres = os.path.splitext(infile)
    print ('splittextres: ', splittextres)
    outfile = os.path.splitext(infile)[0] + '.png'
    print ('outfile: ', outfile)
    
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
            print ('cannot convert ', infile)
```

```python
print ('filename: ', infile)
inimg = Image.open(infile)
plt.imshow(inimg)
```

```python
print ('filename: ', outfile)
outimg = Image.open(outfile)
plt.imshow(outimg)
```
